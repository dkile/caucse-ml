# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## plot the input data

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    distance = np.linalg.norm(feature - center, axis=1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    centroid = np.zeros(shape=np.size(feature, 1))
    isEmpty = np.sum([label_feature == label]) == 0
    if not isEmpty:
        centroid = np.mean(feature[label_feature == label], axis = 0)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return isEmpty, centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label_feature = np.argmin(distance, axis = 1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

```python
def compute_loss(feature, centers, label_feature, num_cluster):
    loss = 0
    for i in range(num_cluster):
        loss += np.sum(np.power(compute_distance(feature[label_feature==i], centers[i]), 2))
    return loss / np.size(feature, 0)
    
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 10000     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
for i in range(number_iteration):
    
    distance_cluster = np.zeros(shape=(number_data, number_cluster))
    
    for cluster_num in range(number_cluster):
        isEmpty, centroid = compute_centroid(feature, label_feature_02, cluster_num)
        if not isEmpty:
            centroid_iteration_02[i][cluster_num] = centroid
        else:
            centroid_iteration_02[i][cluster_num] = centroid_iteration_02[i-1][cluster_num]
        distance_cluster[:,cluster_num] = compute_distance(feature, centroid)
    
    label_feature_02 = compute_label(distance_cluster)
    
    loss = compute_loss(feature, centroid_iteration_02[i], label_feature_02, number_cluster)
    
    loss_iteration_02[i] = loss
#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
