## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    point_x = point[:,0]
    point_y = point[:,1]
    
    # feature = np.stack((np.ones_like(point_x), point_x, np.power(point_x, 2), np.power(point_x, 3), np.power(point_x, 4), np.power(point_x, 5), np.power(point_x, 6), point_y), axis=1)
    feature = np.stack((np.ones_like(point_x), point_x, np.power(point_x, 2), np.power(point_x, 3), np.power(point_x, 4), np.power(point_x, 5), np.power(point_x, 6), point_y, np.power(point_y, 2), np.power(point_y, 3), np.power(point_y, 4), np.power(point_y, 5), np.power(point_y, 6)), axis=1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.matmul(feature, theta.T)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = sigmoid(compute_linear_regression(theta, feature))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    residual = -label * np.log1p(compute_logistic_regression(theta, feature)) - (np.ones_like(label) - label) * np.log1p(1 - compute_logistic_regression(theta, feature))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    loss = (np.sum(compute_residual(theta, feature, label)) / len(label)) + (alpha / 2) * np.inner(theta, theta.T)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    gradient = np.matmul((compute_logistic_regression(theta, feature) - label), feature) / len(label) + alpha*theta

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    bCorrect = (compute_logistic_regression(theta, feature) >= 0.5) == label
    accuracy    = np.sum(bCorrect) / len(label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
number_iteration    = 10000000 # you can change this value as you want 
learning_rate       = 3*1e-3 # you can change this value as you want 
number_feature      = 13 # you can change this value as you want
alpha               = 1e-9 # you can change this value as you want

theta                       = np.zeros(number_feature)
loss_iteration_train        = np.zeros(number_iteration)
loss_iteration_test         = np.zeros(number_iteration)
accuracy_iteration_train    = np.zeros(number_iteration)
accuracy_iteration_test     = np.zeros(number_iteration)
data_train_feature = compute_feature(data_train_point)
data_test_feature = compute_feature(data_test_point)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
 
    theta           = theta - learning_rate*compute_gradient(theta, data_train_feature, data_train_label, alpha)
    loss_train      = compute_loss(theta, data_train_feature, data_train_label, alpha)
    loss_test       = compute_loss(theta, data_test_feature, data_test_label, alpha)
    accuracy_train  = compute_accuracy(theta, data_train_feature, data_train_label)
    accuracy_test   = compute_accuracy(theta, data_test_feature, data_test_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    loss_iteration_train[i]     = loss_train
    loss_iteration_test[i]      = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i]  = accuracy_test

theta_optimal = theta
```
